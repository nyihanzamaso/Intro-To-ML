```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris_data=load_iris()
df=pd.DataFrame(iris_data.data)
df.columns=iris_data.feature_names
df['target']=iris_data.target
```

```python
df['target_names']=df['target'].map({0:'setosa', 1:'versicolor', 2:'virginica'})
```

```python
sns.relplot(x='sepal length (cm)' , y='target', hue='target_names', data=df)
```

```python
sns.pairplot(df, hue='target_names')
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_train, df_test=train_test_split(df, test_size=0.25)
```

```python
x_train=df_train.drop(columns=['target','target_names']).values
```

```python
def single_feature_prediction(petal_length):
    if petal_length<2.5:
        return 0
    elif petal_length<4.8:
        return 1
    else:
        return 2
```

```python
manual_y_prediction=np.array([single_feature_prediction(val) for val in x_train[:,2]])
```

```python
y_train=df_train['target'].values
```

```python
manual_model_accuracy=np.mean(manual_y_prediction == y_train)
```

```python
print(f"manual model accuracy : {manual_model_accuracy * 100:.2f} %")
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score,cross_val_predict
```

```python
import matplotlib.pyplot as plt1
```

```python
def plot_incorrect_predictions(df_predictions, x_axis_feature, y_axis_feature):
    fig, axs= plt1.subplots(2,2, figsize=(10,10))
    axs=axs.flatten()
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue='prediction_labels', data=df_predictions, ax=axs[0])
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue='target_names', data=df_predictions, ax=axs[1])
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue='correct_predictions', data=df_predictions, ax=axs[2])
    axs[3].set_visible(False)
    plt1.show()
```

```python
plot_incorrect_predictions(df_predictions,'petal length (cm)', 'petal width (cm)')
```
